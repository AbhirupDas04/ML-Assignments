```python
from sklearn.model_selection import train_test_split

# Splitting into train, test and val.
train_tmp, val_tmp = train_test_split(train, test_size=0.125, random_state=42)

x_train = train_tmp.drop(columns = ['label']).to_numpy()
x_val = val_tmp.drop(columns = ['label']).to_numpy()
x_test = test.drop(columns = ['label']).to_numpy()

y_train = train_tmp[['label']].to_numpy().ravel() # Ravelling to convert to 1-D array
y_val = val_tmp[['label']].to_numpy().ravel()
y_test = test[['label']].to_numpy().ravel()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import log_loss, accuracy_score

def perform_MLP(activation_fn):
    # MLP Model - setting warm start to true and max_iter to 1, in order to get the val loss after each iteration.
    model = MLPClassifier(random_state = 42, hidden_layer_sizes = (128, 64, 32), max_iter = 1, solver = 'adam', 
                        batch_size = 128, learning_rate = 'constant', learning_rate_init = 2e-5, activation = activation_fn,
                        warm_start = True)
    
    train_loss_arr = []
    val_loss_arr = []

    # Doing it this way since there is no other way to find val loss. 100 epochs of fitting the model, is basically 100 iterations, since
    # we have warm_start on.
    for i in range(100):
        model.fit(x_train, y_train)
        
        y_train_pred = model.predict_proba(x_train) # Using predict_proba and not just predict, since for log loss, we need probability of each class to find final cross entropy loss.
        train_loss = log_loss(y_train, y_train_pred) # Using log_loss since in sklearn, that's the loss that they use to train MLPClassifier.
        train_loss_arr.append(train_loss)

        y_val_pred = model.predict_proba(x_val)
        val_loss = log_loss(y_val, y_val_pred)
        val_loss_arr.append(val_loss)

        # print(f'Train Loss for Activation {activation_fn} at Iter {i+1} -> {train_loss}. Val Loss -> {val_loss}')

    # Plotting results
    plt.plot(range(1, 101), train_loss_arr, label='Training Loss')
    plt.plot(range(1, 101), val_loss_arr, label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title(f'Training and Validation Loss per Epoch for Activation Function {activation_fn}')
    plt.legend()
    plt.show()

    # Finding accuracy on test set.
    y_test_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_test_pred)
    print(f"Accuracy with Activation Function {activation_fn} -> {round(accuracy*100,2)}%.")

activation_fns = ['logistic', 'tanh', 'relu', 'identity']

for activation in activation_fns:
    perform_MLP(activation)
```

Tanh gives us the best accuracy on the test set with 84%. So, we shall use this going forwards.

Part 4.(a), (b) and (c)

a) Since the input and output layers already are of dimension 784, so we set c to be less than 784 to capture important features but not overfit. So, I went for 512 for c. Then for b, I went for 256, and finally for a, I went for 128.

```python
from sklearn.model_selection import train_test_split

# Splitting into train, test and val.
train_tmp, val_tmp = train_test_split(train, test_size=0.125, random_state=42)

x_train = train_tmp.drop(columns = ['label'])
x_val = val_tmp.drop(columns = ['label'])
x_test = test.drop(columns = ['label'])

# Y here will be just a copy of x, since we are trying to regenerate the original image.
y_train = x_train.copy(deep = True)
y_val = x_val.copy(deep = True)
y_test = x_test.copy(deep = True)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPRegressor

# Making model with activation relu. Putting warm start = True and max_iter = 1 to get val loss after each iteration, otherwise no other way
# to do so.
model_relu = MLPRegressor(random_state = 42, hidden_layer_sizes = (512, 256, 128, 256, 512), activation = 'relu', solver = 'adam', 
                          learning_rate = 'constant', learning_rate_init = 2e-5, warm_start = True, max_iter = 1)

# Activation is 'identity'
model_identity = MLPRegressor(random_state = 42, hidden_layer_sizes = (512, 256, 128, 256, 512), activation = 'identity', solver = 'adam', 
                              learning_rate = 'constant', learning_rate_init = 2e-5, warm_start = True, max_iter = 1)
```

```python
from sklearn.metrics import root_mean_squared_error

def perform_MLP(model, activation):
    train_loss_arr = []
    val_loss_arr = []

    # Doing it this way since there is no other way to find val loss. 100 epochs of fitting the model, is basically 100 iterations, since
    # we have warm_start on.
    for i in range(100):
        model.fit(x_train, y_train)
        
        y_train_pred = model.predict(x_train) # Finding the regenerated image.
        train_mse = root_mean_squared_error(y_train, y_train_pred) ** 2 # MSE = RMSE^2
        train_loss_arr.append(train_mse)

        y_val_pred = model.predict(x_val)
        val_mse = root_mean_squared_error(y_val, y_val_pred) ** 2
        val_loss_arr.append(val_mse)

        print(f'Train Loss for Activation {activation} at Iter {i+1} -> {train_mse}. Val Loss -> {val_mse}')

    # Plotting results
    plt.plot(range(1, 101), train_loss_arr, label='Training Loss')
    plt.plot(range(1, 101), val_loss_arr, label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title(f'Training and Validation Loss per Epoch for Activation Function {activation}')
    plt.legend()
    plt.show()

    return model

model_relu = perform_MLP(model_relu, 'relu')
model_identity = perform_MLP(model_identity, 'identity')
```

```python
# Saving the models by using the pickle module
with open('model_relu.pkl', 'wb') as file:
    pickle.dump(model_relu, file)

with open('model_identity.pkl', 'wb') as file:
    pickle.dump(model_identity, file)
```

Part 5

```python
train = pd.read_csv('./fashion-mnist_train.csv')
test = pd.read_csv('./fashion-mnist_test.csv')

train_tmp = train.head(8000)
test_tmp = test.head(2000)

x_train = train_tmp.drop(columns = ['label'])
x_test = test_tmp.drop(columns = ['label'])

y_train = train_tmp[['label']].to_numpy().ravel() # Ravelling to convert to 1-D array
y_test = test_tmp[['label']].to_numpy().ravel()
```

```python
x_train_relu = x_train.copy(deep = True).to_numpy()
x_train_identity = x_train.copy(deep = True).to_numpy()

x_test_relu = x_test.copy(deep = True)
x_test_identity = x_test.copy(deep = True)
```

```python
# For ReLU
for i in range(3):
    train_tmp = np.dot(x_train_relu, model_relu.coefs_[i]) + model_relu.intercepts_[i] # Train features
    test_tmp = np.dot(x_test_relu, model_relu.coefs_[i]) + model_relu.intercepts_[i] # Test features
    x_train_relu = np.maximum(0, train_tmp)  # ReLU
    x_test_relu = np.maximum(0, test_tmp)
```

```python
# For Identity (Activation is identity, so after multiplying by weights and adding bias, we directly put in to next layer.)
for i in range(3):
    x_train_identity = np.dot(x_train_identity, model_relu.coefs_[i]) + model_relu.intercepts_[i] # Train features
    x_test_identity = np.dot(x_test_identity, model_relu.coefs_[i]) + model_relu.intercepts_[i] # Test features
```

```python
# Classifier RELU
model = MLPClassifier(random_state = 42, hidden_layer_sizes = (128, 128), max_iter = 200, solver = 'adam', activation = 'relu', 
                      learning_rate = 'constant', learning_rate_init = 2e-5)
model.fit(x_train_relu, y_train)

# Finding accuracy on test set.
y_test_pred = model.predict(x_test_relu)
accuracy = accuracy_score(y_test, y_test_pred)
print(f"Accuracy with Activation Function Relu -> {round(accuracy*100, 2)}%.")
```

```python
# Classifier Identity
model = MLPClassifier(random_state = 42, hidden_layer_sizes = (128, 128), max_iter = 200, solver = 'adam', activation = 'identity', 
                      learning_rate = 'constant', learning_rate_init = 2e-5)
model.fit(x_train_identity, y_train)

# Finding accuracy on test set.
y_test_pred = model.predict(x_test_identity)
accuracy = accuracy_score(y_test, y_test_pred)
print(f"Accuracy with Activation Function Identity -> {round(accuracy*100, 2)}%.")
```

This method still is a decent classifier(8.55 % less than in Part 2) since the hidden layers capture crucial information with regards to the details of the image, like its shape, intensity, the nature of the object, otherwise it wouldn't be able to reconstruct the image at all. Because of this, when we use the hidden layer's representation/embedding, the classification works pretty well, as those features itself are used to identify an object.

So, to summarise, reasons:-

1) Hidden Layers capture crucial info like shape, borders, gradients, type of object, etc.
2) These details themselves are used in the classifier in Part 2, where we slowly reduced the size of the hidden layer.
3) So, during classification, since they carry the info regarding the features, they are a pretty decent classifier.
